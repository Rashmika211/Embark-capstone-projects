#Cleaing the Dataset

```python
y_train = train['SalePrice']
test_id = test['Id']

train.drop(['SalePrice','Id'],axis=1,inplace=True)
test.drop('Id',axis=1,inplace=True)
```

```python
#combining train and test data
all_data = pd.concat([train,test],axis=0)
```

```python
sns.heatmap(all_data.isnull(),cbar=False)
```

```python
#Filling categorical values
all_data.select_dtypes(exclude = object)
```

```python
#Values that should be Categorical
cat_col1 = ['MSSubClass','OverallCond','OverallQual']
all_data[cat_col1] = all_data[cat_col1].astype(str)
for i in cat_col1:
  all_data[i] = all_data[i].fillna(all_data[i].mode()[0])
```

```python
date_col = ['YrSold','MoSold','YearBuilt','YearRemodAdd','GarageYrBlt']

for i in date_col:
  all_data[i] = all_data[i].fillna(all_data[i].mode()[0])
```

```python
all_data.select_dtypes(include = object).columns
```

```python
#features with blank values having meaning
cat_col2 = ["Alley","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1",
            "BsmtFinType2","FireplaceQu","GarageType","GarageFinish",
            "GarageQual", "GarageCond","PoolQC", "Fence","MiscFeature"]
#replacing NA by None
for i in cat_col2:
  all_data[i] = all_data[i].fillna("None")
```

```python
all_data.select_dtypes(include=object).isna().sum().sort_values(ascending=False)[all_data.isna().sum()!=0]
```

```python
cat_col = all_data.select_dtypes(include=object).columns
for i in cat_col:
  all_data[i] = all_data[i].fillna(all_data[i].mode()[0])
```

```python
all_data.select_dtypes(include=object).isna().sum().sort_values(ascending=False)[all_data.isna().sum()!=0]
```

```python
#Filling numerical values
all_data.select_dtypes(exclude=object).isna().sum().sort_values(ascending=False)[all_data.isna().sum()!=0]
```

```python
num_col = all_data.select_dtypes(exclude=object).columns
for i in num_col:
  all_data[i] = all_data[i].fillna(all_data[i].mean())
```

```python
sns.heatmap(all_data.isnull(),cbar=False)
```

#Feature transformation

```python
from scipy.stats import skew
skewed_feats = all_data[num_col].apply(lambda x: abs(skew(x))).sort_values(ascending=False)

skewed_feats
```

```python
#applying log transformation 
for col in skewed_feats.index:
   all_data[col] = np.log1p(all_data[col])
```

```python
all_data[num_col].apply(lambda x: abs(skew(x))).sort_values()
```

#Model Selection

```python
from pycaret.regression import setup, compare_models
```

```python
M = setup(data = pd.concat([train, log_ytrain],axis=1), target = 'SalePrice')
```

```python
compare_models()
```

#Model Training

```python
from catboost import CatBoostRegressor
from sklearn.linear_model import BayesianRidge, Ridge, OrthogonalMatchingPursuit,HuberRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
models = {
    "catboost": CatBoostRegressor(verbose=0),
    "br": BayesianRidge(),
    "lightgbm": LGBMRegressor(),
    "ridge": Ridge(),
    "omp": OrthogonalMatchingPursuit(),
    "gbr" : GradientBoostingRegressor(),
    "huber":HuberRegressor()

}
```

```python
for i, model in models.items():
    model.fit(train, log_ytrain)
    print(i + " trained.")
```

```python
rmsle_model = {}

for i, model in models.items():
  rmsle = np.exp(np.sqrt(-cross_val_score(model,train,log_ytrain,scoring='neg_mean_squared_log_error', cv = 10)))
  rmsle_model[i] = rmsle
```

```python
rmsle_model
```

```python
for i, rmsle in rmsle_model.items():
    print("----------\n" + i)
    print(np.mean(rmsle))
    print(np.std(rmsle))
```

```python
final_predictions = (
    0.4 * np.exp(models['catboost'].predict(test)) +
    0.2 * np.exp(models['huber'].predict(test)) +
    0.2 * np.exp(models['br'].predict(test)) +
    0.1 * np.exp(models['gbr'].predict(test)) +
    0.1 * np.exp(models['lightgbm'].predict(test))
)
```

#Submission

```python
submission = pd.DataFrame({'Id':test_id,'SalePrice':final_predictions})
print(submission)
submission.to_csv('submission_file',index=False)
```
